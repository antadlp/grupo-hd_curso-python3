```python
def config_md(**kwargs):
    
    MDORG_FILE = kwargs.get("mdorg_file", "mdorg.dat")
    
    f = open(MDORG_FILE, 'r')
    
    N = int(f.readline().split()[0])
    
    L = f.readline().split()
    Lx = float(L[0])
    Ly = float(L[1])
    Lz = float(L[2])
    
    rx = []
    ry = []
    rz = []
    
    vx = []
    vy = []
    vz = []
    
    for line in f:
        
        sp = line.split()
        
        rx.append(sp[0])
        ry.append(sp[1])
        rz.append(sp[2])
        
        vx.append(sp[3])
        vy.append(sp[4])
        vz.append(sp[5])
    
    f.close()
    
    dic = {}
    dic['rx'] = np.array(rx).astype(np.float)
    dic['ry'] = np.array(ry).astype(np.float)
    dic['rz'] = np.array(rz).astype(np.float)
    
    dic['vx'] = np.array(vx).astype(np.float)
    dic['vy'] = np.array(vy).astype(np.float)
    dic['vz'] = np.array(vz).astype(np.float)
    
    dic['N'] = N
    
    dic['Lx'] = Lx
    dic['Ly'] = Ly
    dic['Lz'] = Lz

    
    return dic


        
    
```
